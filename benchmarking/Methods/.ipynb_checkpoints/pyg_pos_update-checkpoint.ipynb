```python
# Import Packages
#============================================
import torch 
from torch_geometric.datasets import QM9
from rdkit import Chem
from rdkit.Geometry import Point3D
from rdkit.Chem import rdMolTransforms
from torch_geometric.data import Data
from equiformer import EquiformerQM9
from torch_geometric.data import Batch
import torch.nn.functional as F
import numpy as np
import math
#=====================================
```

```python
#============================================
# Converting Pyg into RDKit mol
#============================================
def pyg_qm9_to_rdkit_mol(data, smiles: str):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None :
        raise ValueError("RDKit failed")

    mol = Chem.AddHs(mol)
    n_mol = mol.GetNumAtoms()
    n_pos = int(data.pos.size(0))
    if n_mol != n_pos:
        raise ValueError(f"Atom count mismatch: RDKit mol has {n_mol} atoms, PyG data.pos has {n_pos} atoms.\n"
            "This usually means atom ordering or explicit-H handling differs.")

    conf = Chem.Conformer(n_mol)
    pos = data.pos.detach().cpu()

    for i in range(n_mol):
        x, y, z = pos[i].tolist()
        conf.SetAtomPosition(i, Point3D(float(x), float(y), float(z)))

    mol.RemoveAllConformers()
    mol.AddConformer(conf, assignId = True)
    return mol
```

```python
#===============================================
# Get all the torsion angles of one Molecule
#===============================================
def get_all_torsion_angles(mol):

    if mol.GetNumConformers() == 0:
        raise ValueError("Molecule has no 3D coordinates attached")

    conf = mol.GetConformer()
    torsions = []
    for bond in mol.GetBonds():
        j = bond.GetBeginAtomIdx()
        k = bond.GetEndAtomIdx()
        
        nbrs_j = [a.GetIdx() for a in mol.GetAtomWithIdx(j).GetNeighbors() if a.GetIdx() != k]
        nbrs_k = [a.GetIdx() for a in mol.GetAtomWithIdx(k).GetNeighbors() if a.GetIdx() != j]

        if not nbrs_j or not nbrs_k:
            continue

        for i in nbrs_j :
            for l in nbrs_k:
                angle = rdMolTransforms.GetDihedralDeg(conf, int(i), int(j), int(k), int(l))
                torsions.append({
                    "atoms": (int(i), int(j), int(k), int(l)),
                    "bond": (int(j), int(k)),
                    "angle_deg": float(angle),
                })

    return torsions
```

```python
#==================================================================
# Set the torsion angles 
#==================================================================
def set_torsion_angle(mol, i, j, k, l, new_angle , conf_id = 0):
    
    mol_copy = Chem.Mol(mol)     # Coppy as SetDihedralDeg function updates mol in place
    
    if mol_copy.GetNumConformers()== 0:
        raise ValueError("Molecule has no conformer")
    conf = mol_copy.GetConformer(conf_id)
    rdMolTransforms.SetDihedralDeg(conf, int(i), int(j), int(k), int(l), float(new_angle))

    return mol_copy
```

```python
#=============================================================
# Converting the new Molecule from RDKit to Pyg 
#=============================================================
def rdkit_to_pyg_update_pos(original_data, mol, conf_id=0):

    if mol.GetNumConformers() == 0:
        raise ValueError("RDKit mol has no conformer.")

    conf = mol.GetConformer(conf_id)
    n_atoms = mol.GetNumAtoms()

    if n_atoms != original_data.pos.size(0):
        raise ValueError(
            f"Atom mismatch! RDKit has {n_atoms} atoms but PyG has {original_data.pos.size(0)}"
        )

    new_pos = torch.tensor(
        [[conf.GetAtomPosition(i).x,
          conf.GetAtomPosition(i).y,
          conf.GetAtomPosition(i).z] for i in range(n_atoms)],
        dtype=original_data.pos.dtype,
        device=original_data.pos.device
    )

    new_data = original_data.clone()
    new_data.pos = new_pos

    return new_data
```

```python
#=====================================
# Import data 
#=====================================
dataset = QM9(root= "data/QM9")

idx = 18             # Choosing the molecule from the data
data = dataset[idx]
smiles = dataset.smiles[idx]
mol = pyg_qm9_to_rdkit_mol(data, smiles)

print(data)
print("SMILES:", smiles)
print("num atoms:", data.pos.size(0))
print("RDKit atoms: ", mol.GetNumAtoms(), "conformers: ", mol.GetNumConformers())
```

```python
#================================================
# Run all Functions 
#================================================
torsions = get_all_torsion_angles(mol)
t = torsions[0]                # Choosing the torsion 
i, j, k, l = t["atoms"]
mol_new = set_torsion_angle(mol, i, j, k, l, 160.0)
new_data = rdkit_to_pyg_update_pos(data, mol_new)
print("old_mol:", rdMolTransforms.GetDihedralDeg(mol.GetConformer(), i, j, k, l))
print("new_mol:", rdMolTransforms.GetDihedralDeg(mol_new.GetConformer(), i, j, k, l))
```

```python
#======================================================
# Device - Checkpoint - Calling the model 
#======================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
checkpoint = torch.load("/project/IZZY/MolRepres/benchmarking/Methods/checkpoints_Equiformernew/best_model.pt", map_location=device)
print(checkpoint.keys())

# Equiformer 
model = EquiformerQM9().to(device)
model.load_state_dict(checkpoint["model_state_dict"])
```

```python
#======================================================
# Feed the new and original Molecule into the model
#======================================================

# Original Molecule
with torch.no_grad():
    m1 = model(Batch.from_data_list([data]).to(device))

print("Original_Molecule:",m1)


# New Molecule
with torch.no_grad():
    m2 = model(Batch.from_data_list([new_data]).to(device))

print("New_Molecule:",m2)
```

```python
#==========================================================
# Embedding Space
#==========================================================

# Orig Mol
with torch.inference_mode():
    batch = Batch.from_data_list([data]).to(device)
    z1 = model.encode(batch).cpu()

# New Mol
with torch.inference_mode():
    batch = Batch.from_data_list([new_data]).to(device)
    z2 = model.encode(batch).cpu()

print("Embedding_Space_Org: ", z1)
print("Embedding_Space_New: ", z2)
```

```python
cos_sim = F.cosine_similarity(z1, z2, dim=1)
print(cos_sim)
```

```python
def rdkit_mol_to_pyg(mol:Chem.Mol,
                    conf_id: int = 0, 
                    require_conformer:bool = True,
                    embed_if_missing: bool =True,
                    add_hs:bool = False,
                    include_edge_attr: bool = False,
                    include_x: bool = False,
                    y = None, 
                    **meta, 
                    ) -> Data:
    if mol is None:
        raise ValueError("mol is None")

    mol = Chem.Mol(mol)

    if add_hs:
        mol = Chem.AddHs(mol)

    if mol.GetNumConformers() == 0:
        if require_conformer:
            if not embed_if_missing:
                raise ValueError ("Mol has no conformer and embed_if_missing = False.")

            params = AllChem.ETKDGv3()
            params.randomSeed = 0
            ok = AllChem.EmbedMolecule(mol, params)
            if ok != 0:
                raise ValueError("RDKit failed to embed a conformer (ETKDG).")
            try:
                AllChem.UFFOptimizeMolecule(mol, maxIters = 200)
            except Exception:
                pass
        else :
            conf_id = None

    n = mol.GetNumAtoms()
    z = torch.tensor([mol.GetAtomWithIdx(i).GetAtomicNum()for i in range (n)], dtype = torch.long)

    if conf_id is not None:
        conf = mol.GetConformer(conf_id)
        pos = torch.tensor(
            [[conf.GetAtomPosition(i).x, conf.GetAtomPosition(i).y, conf.GetAtomPosition(i).z] for i in range(n)],
            dtype = torch.float32,
        )
    else:
        pos = None

    row, col = [], []
    edge_attr = []

    for bond in mol.GetBoond():
        a, b = bond.GetBeginAtomIdx(), bond.GetEndAtomIdx

        row += [a, b]
        col += [b, a]

        if include_edge_attr:
            onehot = [0.0, 0.0, 0.0, 0.0]
            bt = Boon_types.get(bond.GetBondType(), None)
            if bt is not None:
                onehot[bt] = 1.0

            edge_attr += [onehot, onehot]
    edge_index = torch.tensor([row, col], dtype = torh.long)
    data_kwargs = {"z": z, "edge_index": edge_index}
    if pos is not None:
        data_kwargs["pos"]= pos
    data = Data(**data_kwargs)

    if include_edge_attr:
        data.edge_attr = torch.tensor(edge_attr, dtype = torch.float32)
    if include_x:
        x = []
        for i in range(n):
            a = mol.GetAtomWithIdx(i)
            x.append([
                float(a.GetAtomicNum()),
                float(a.GetDegree()),
                float(a.GetFormalCharge()),
                1.0 if a.GetIsAromatic() else 0.0,
            ])
        data.x = torch.tensor(x, dtype = torch.float32)

    if y is not None:
        data.y = y

    for k, v in meta.items():
        setattr(data, k, v)

    return data
```
