```python
# _____________GCN Model- Data-QM9____________

import numpy as np
import torch 
import torch.nn as nn
from torch_geometric.nn import GCNConv
import torch.nn.functional as F
from torch_geometric.datasets import QM9
from torch_geometric.loader import DataLoader
from sklearn.preprocessing import StandardScaler
from ase.units import Hartree, eV, Bohr, Ang
from torch_geometric.nn import global_mean_pool

# Load QM9 dataset 
dataset = QM9(root='data/QM9')

# First 19 target properties
target_dim = 7

# Define the correct unit conversions 
def get_qm9_conversions():
    """Unit conversions for QM9 properties"""
    conversions = [
        
        Hartree / eV,           # 7: U₀ (eV)- Internal energy at 0 K
        
    ]
    return conversions

# Get raw targets and apply unit conversions (only first 6)
y_original = dataset.data.y[:, :target_dim].cpu().numpy()
conversions = get_qm9_conversions()

# Apply unit conversions
y_converted = y_original.copy()
for i in range(target_dim):
    y_converted[:, i] = y_original[:, i] * conversions[i]

# Store statistics for normalization using training set 
train_mean = y_converted[:10000].mean(axis=0)  # Use training set statistics
train_std = y_converted[:10000].std(axis=0)

print("Training set statistics (19 targets):")
property_names = ['μ (D)', 'α (Ang³)', 'ε_HOMO (eV)', 'ε_LUMO (eV)', 'Δε (eV)', '⟨R²⟩ (Ang²)', 'ZPVE (ev)', 'U₀ (eV)', 'U (eV)', 'H (eV)', 'G (eV)',
                 'c_v (cal/(mol⁻k))', 'U₀_atom (eV)', 'U_atom (eV)', 'H_atom (eV)', 'G_atom (eV)', 'A (GHz)', 'B (GHz)', 'C (GHz)']

for i, name in enumerate(property_names):
    print(f"  {name:20s}: mean={train_mean[i]:8.3f}, std={train_std[i]:8.3f}")

# Normalize using training statistics
y_normalized = (y_converted - train_mean) / train_std

# Replace only first 6 targets in dataset
dataset.data.y= torch.tensor(y_normalized, dtype=torch.float)

# Split dataset (same splits)
train_dataset = dataset[:10000]
val_dataset = dataset[10000:11000]
test_dataset = dataset[11000:12000]

# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Simple GNN model
class GCNModel(nn.Module):
    def __init__(self, target_dim=7):
        super().__init__()
        self.conv1 = GCNConv(11, 64)  # 11 node features in QM9
        self.conv2 = GCNConv(64, 64)
        self.conv3 = GCNConv(64, 64)
        self.fc1 = nn.Linear(64, 32)
        self.fc2 = nn.Linear(32, target_dim)
        self.dropout = nn.Dropout(0.5)
        
    def forward(self, x, edge_index, batch):
        x = F.relu(self.conv1(x, edge_index))
        x = self.dropout(x)
        x = F.relu(self.conv2(x, edge_index))
        x = self.dropout(x)
        x = F.relu(self.conv3(x, edge_index))
        
        # Global mean pooling
        x = global_mean_pool(x, batch)
        # MLP layers
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

# Initialize model, optimizer
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = GCNModel(target_dim=target_dim).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

def train():
    model.train()
    total_loss = 0
    for batch in train_loader:
        batch = batch.to(device)
        optimizer.zero_grad()
        pred = model(batch.x, batch.edge_index, batch.batch)
        # Train on normalized targets 
        loss = F.mse_loss(pred, batch.y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(train_loader)

def test(loader):
    model.eval()
    total_mse = 0
    total_mae = 0  # Just track overall MAE
    count = 0
    
    with torch.no_grad():
        for batch in loader:
            batch = batch.to(device)
            pred = model(batch.x, batch.edge_index, batch.batch)

            # Calculate loss on normalized data during training
            mse = F.mse_loss(pred, batch.y).item()
            total_mse += mse
            count += 1
    
    avg_mse = total_mse / count
    return avg_mse

def evaluate_with_proper_units(loader):
    """Evaluate model and return MAE in original units"""
    model.eval()
    total_mae_per_attr = torch.zeros(target_dim, device=device)
    count = 0
    
    with torch.no_grad():
        for batch in loader:
            batch = batch.to(device)
            pred_normalized = model(batch.x, batch.edge_index, batch.batch)
            target_normalized = batch.y

            # Denormalize for proper MAE calculation
            pred_denorm = pred_normalized * torch.tensor(train_std, device=device, dtype=torch.float) + torch.tensor(train_mean, device=device, dtype=torch.float)
            target_denorm = target_normalized * torch.tensor(train_std, device=device, dtype=torch.float) + torch.tensor(train_mean, device=device, dtype=torch.float)
            
            # Calculate MAE in original units
            mae_per_attr = (pred_denorm - target_denorm).abs().mean(dim=0)
            total_mae_per_attr += mae_per_attr
            count += 1
    
    avg_mae_per_attr = (total_mae_per_attr / count).cpu().numpy()
    return avg_mae_per_attr

# Training loop
print("\nStarting training...")
for epoch in range(1, 51):
    train_loss = train()
    val_mse = test(val_loader)  # Use normalized data for validation during training
    print(f"Epoch {epoch:03d}, Train Loss: {train_loss:.4f}, Val MSE: {val_mse:.4f}")

# Final evaluation with proper units
print("\nFinal Evaluation (Proper Units):")
test_mae = evaluate_with_proper_units(test_loader)
print(f"MAE per attribute (GCN Model):")

for i, (mae, name) in enumerate(zip(test_mae, property_names)):
    print(f"  {i:2d}. {name:20s}: {mae:8.3f}")
```

```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.datasets import QM9
from torch_geometric.nn import GCNConv, global_mean_pool
from torch_geometric.loader import DataLoader
from torch.optim import Adam
from torch.optim.lr_scheduler import StepLR
from tqdm import tqdm
from ase.units import Hartree, eV, Bohr, Ang
from torch.utils.tensorboard import SummaryWriter
import numpy as np
from dig.threedgraph.method import SphereNet

# QM9 Targerts
PROPERTY_NAMES = [
    'μ (D)', 'α (Ang³)', 'ε_HOMO (eV)', 'ε_LUMO (eV)', 'Δε (eV)', '⟨R²⟩ (Ang²)',
    'ZPVE (eV)', 'U₀ (eV)', 'U (eV)', 'H (eV)', 'G (eV)', 'c_v', 'U₀_atom',
    'U_atom', 'H_atom', 'G_atom', 'A (GHz)', 'B (GHz)', 'C (GHz)'
]

TARGET_PROPERTY = 'μ (D)'   # Change the target here with replacing it with another from PROPERTY_NAMES
TARGET_INDEX = PROPERTY_NAMES.index(TARGET_PROPERTY)

# Device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Using device:", device)

# QM9 conversion factors 
def get_qm9_conversions_tensor(device):
    return torch.tensor([
        1.0, Bohr**3 / Ang**3, Hartree / eV, Hartree / eV, Hartree / eV,
        Bohr**2 / Ang**2, Hartree / eV, Hartree / eV, Hartree / eV, Hartree / eV,
        Hartree / eV, 1.0, 1.0, Hartree / eV, Hartree / eV, Hartree / eV,
        1.0, 1.0, 1.0
    ], dtype=torch.float, device=device)

# Load dataset and split
dataset = QM9(root='data/QM9')
num_mols = len(dataset)
print(f"Total QM9 molecules: {num_mols}")

train_size, valid_size = 110000, 10000
assert train_size + valid_size < num_mols, "Split sizes too large for dataset"

perm = torch.randperm(num_mols, generator=torch.Generator().manual_seed(42))
train_idx = perm[:train_size]
valid_idx = perm[train_size:train_size+valid_size]
test_idx  = perm[train_size+valid_size:]

# Normalize a single target 
# Extract raw target vector 
y_raw = dataset.data.y[:, TARGET_INDEX].clone().cpu() 

conversions_cpu = get_qm9_conversions_tensor(device='cpu')
y_conv_cpu = y_raw * conversions_cpu[TARGET_INDEX].cpu()

train_y_cpu = y_conv_cpu[train_idx.cpu()]  # indices are CPU tensor
train_mean_cpu = float(train_y_cpu.mean().item())
train_std_cpu  = float(train_y_cpu.std().item())
if train_std_cpu == 0:
    train_std_cpu = 1.0

# Normalize and overwrite dataset.data.y with shape [N,1] on CPU
y_norm_cpu = (y_conv_cpu - train_mean_cpu) / train_std_cpu
dataset.data.y = y_norm_cpu.unsqueeze(1).to(torch.float)

# Save normalization constants for denorm later
norm_stats = {'mean': train_mean_cpu, 'std': train_std_cpu}
print("Normalization (mean, std):", norm_stats)

# Build splits 
train_dataset = dataset[train_idx]
valid_dataset = dataset[valid_idx]
test_dataset  = dataset[test_idx]

# Model 
in_channels = dataset.num_node_features
print("Node feature dim:", in_channels)

class GCNModel(nn.Module):
    def __init__(self, in_dim, hidden_dim=128, dropout=0.3):
        super().__init__()
        self.conv1 = GCNConv(in_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, hidden_dim)
        self.conv3 = GCNConv(hidden_dim, hidden_dim)
        self.fc1 = nn.Linear(hidden_dim, 64)
        self.fc2 = nn.Linear(64, 1)
        self.dropout = nn.Dropout(dropout)

    def forward(self, batch):
        x, edge_index, batch_atm = batch.x, batch.edge_index, batch.batch
        x = F.relu(self.conv1(x, edge_index))
        x = F.relu(self.conv2(x, edge_index))
        x = F.relu(self.conv3(x, edge_index))
        x = global_mean_pool(x, batch_atm)   # The global mean pool ing takes the mean of all atom embeddings in each molecule´s graph "one feat vector per molecule"
        x = F.relu(self.fc1(x)) # Decoder : it transfomrs the pooled molecular embedding into the predicted target value 
        x = self.dropout(x)
        return self.fc2(x)


# Train 
def train_epoch(model, loader, optimizer, device):
    model.train()
    total_loss = 0.0
    for batch in tqdm(loader, desc="Train", leave=False):
        batch = batch.to(device)
        optimizer.zero_grad()
        pred = model(batch)                
        target = batch.y                   
        loss = F.mse_loss(pred, target)    
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(loader)

@torch.no_grad()
def evaluate(model, loader, device, norm_mean, norm_std):
    model.eval()
    preds = []
    targets = []
    for batch in tqdm(loader, desc="Eval", leave=False):
        batch = batch.to(device)
        out = model(batch).detach().cpu()    # normalized predictions [B,1]
        t   = batch.y.detach().cpu()         # normalized targets [B,1]
        preds.append(out)
        targets.append(t)
    preds = torch.cat(preds, dim=0)
    targets = torch.cat(targets, dim=0)
    # Denormalize 
    preds_den = preds * norm_std + norm_mean
    targets_den = targets * norm_std + norm_mean
    mae = torch.mean(torch.abs(preds_den - targets_den)).item()
    return mae


# Trainig Loop
def main():
    # Hyperparams
    epochs = 1000
    batch_size = 128
    vt_batch_size = 256
    lr = 1e-3
    lr_decay_step = 50
    lr_decay_factor = 0.5
    weight_decay = 0.0
    save_dir = 'checkpoints_gcn_single'
    log_dir = 'logs_gcn_single'
    os.makedirs(save_dir, exist_ok=True)
    os.makedirs(log_dir, exist_ok=True)
    writer = SummaryWriter(log_dir=log_dir)

    # Dataloaders 
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader   = DataLoader(valid_dataset, batch_size=vt_batch_size, shuffle=False)
    test_loader  = DataLoader(test_dataset, batch_size=vt_batch_size, shuffle=False)

    #model = GCNModel(in_channels).to(device)
    model = SphereNet(energy_and_force=False, cutoff=5.0, num_layers=4,
        hidden_channels=128, out_channels=1, int_emb_size=64,
        basis_emb_size_dist=8, basis_emb_size_angle=8, basis_emb_size_torsion=8, out_emb_channels=256,
        num_spherical=3, num_radial=6, envelope_exponent=5,
        num_before_skip=1, num_after_skip=2, num_output_layers=3, use_node_features=True
        ).to(device)
    optimizer = Adam(model.parameters(), lr=lr, weight_decay=weight_decay)
    scheduler = StepLR(optimizer, step_size=lr_decay_step, gamma=lr_decay_factor)

    best_val = float('inf')
    best_test = float('inf')

    print("#params:", sum(p.numel() for p in model.parameters()))
    print("Training for target:", TARGET_PROPERTY)

    for epoch in range(1, epochs + 1):
        print(f"\n=== Epoch {epoch} ===")
        train_loss = train_epoch(model, train_loader, optimizer, device)
        val_mae = evaluate(model, val_loader, device, norm_stats['mean'], norm_stats['std'])
        test_mae = evaluate(model, test_loader, device, norm_stats['mean'], norm_stats['std'])

        print(f"Train loss (MSE): {train_loss:.6f} | Val MAE: {val_mae:.6f} | Test MAE: {test_mae:.6f}")
        writer.add_scalar('train_loss', train_loss, epoch)
        writer.add_scalar('val_mae', val_mae, epoch)
        writer.add_scalar('test_mae', test_mae, epoch)

        if val_mae < best_val:
            best_val = val_mae
            best_test = test_mae
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'best_val': best_val
            }, os.path.join(save_dir, 'best_model.pt'))
            print("Saved best model checkpoint.")

        scheduler.step()

    writer.close()
    print("\nFinished training.")
    print("Best validation MAE:", best_val)
    print("Test MAE at best val:", best_test)

if __name__ == "__main__":
    main()
```
