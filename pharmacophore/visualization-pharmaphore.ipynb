```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/project/IZZY/molecular-representation/pharmacophore/results_pharmaco_AUROC.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


methods = df.columns[1:]  
x = np.arange(len(df["Protein"]))
width = 0.1

plt.figure(figsize=(12, 6))

for i, method in enumerate(methods):
    plt.bar(x + i * width, df[method], width, label=method)

plt.xlabel("Protein")
plt.ylabel("AUROC")
plt.title("Pharmacophore Benchmarking per Protein")

plt.xticks(x + width * (len(methods)-1)/2, df["Protein"], rotation=50)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))

for i, method in enumerate(methods):
    plt.bar(x + i * width, df[method], width, label=method)

plt.xlabel("Target Protein")
plt.ylabel("Matching Score")
plt.title("Comparison of Pharmacophore Matching Methods")

plt.xticks(x + width * (len(methods)-1)/2, df["Protein"], rotation=60, ha="right")
plt.legend()

plt.grid(True, axis="y")
plt.tight_layout()
plt.show()
```

```python
df_percent = df.copy()

methods = df.columns[1:]  # skip "Protein"

df_percent[methods] = df_percent[methods] * 100
```

```python
import numpy as np
import matplotlib.pyplot as plt

methods = df_percent.columns[1:]
x = np.arange(len(df_percent["Protein"]))
width = 0.2

plt.figure(figsize=(14, 6))

for i, method in enumerate(methods):
    plt.bar(x + i * width, df_percent[method], width, label=method)

plt.xlabel("Target Protein")
plt.ylabel("AUCROC")
plt.title("Pharmacophore Matching Performance")

plt.xticks(x + width * (len(methods)-1)/2, df_percent["Protein"], rotation=50, ha="right")
plt.legend()

plt.grid(True, axis="y")
plt.tight_layout()
plt.show()
```
